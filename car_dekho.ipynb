```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
car_dekhoo=pd.read_csv('car_dekho.csv')
car_dekhoo
```

```python
car_dekhoo.drop(columns=['seller_type'],inplace=True)
car_dekhoo
```

```python
car_dekhoo['year']=2024-car_dekhoo['year']
car_dekhoo
```

```python
car_dekhoo['company']=car_dekhoo['name'].apply(lambda x:x.split()[0])
car_dekhoo
```

```python
car_dekhoo.drop(columns=['name'],inplace=True)
car_dekhoo
```

```python
car_dekhoo=pd.get_dummies(car_dekhoo,drop_first=True)
car_dekhoo
```

```python

x=car_dekhoo.drop(columns=['selling_price'])
y=car_dekhoo['selling_price']
```

```python
car=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)
car.fit(x_train,y_train)
```

```python
y_pred=car.predict(x_test)
```

```python
pd.DataFrame({'actual':y_test,'predicted':y_pred})
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
